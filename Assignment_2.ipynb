# Q1) Prediction of fraudulent transactions

A bank monitors credit-card transactions for possible fraud.

For each transaction, let the sample space be

$$
\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\},
$$

where:

- $\omega_1$: low-risk customer, legitimate transaction,
- $\omega_2$: low-risk customer, fraudulent transaction,
- $\omega_3$: high-risk customer, legitimate transaction,
- $\omega_4$: high-risk customer, fraudulent transaction.

Assume the probabilities are

$$
P(\{\omega_1\})=0.50,\qquad
P(\{\omega_2\})=0.10,\qquad
P(\{\omega_3\})=0.20,\qquad
P(\{\omega_4\})=0.20.
$$

Let $X:\Omega\to\mathbb{R}$ denote the fraud indicator:

$$
X(\omega_1)=0,\qquad
X(\omega_2)=1,\qquad
X(\omega_3)=0,\qquad
X(\omega_4)=1.
$$

Thus $X=1$ means fraudulent, $X=0$ means legitimate.

Now suppose the bank observes only the customer's risk class. Define $Y:\Omega\to\mathbb{R}$ by

$$
Y(\omega_1)=0,\qquad
Y(\omega_2)=0,\qquad
Y(\omega_3)=1,\qquad
Y(\omega_4)=1.
$$

Thus:

- $Y=0$ means "low-risk customer",
- $Y=1$ means "high-risk customer."

---

## 1. Probability space $(\Omega, \mathfrak{F}, P)$

We have

$$
\Omega = \{\omega_1, \omega_2, \omega_3, \omega_4\}, \quad \mathfrak{F} = \mathfrak{P}(\Omega) \text{ (power set)}.
$$

Probability measure $P$ is given by

$$
P(\omega_1)=0.50,\; P(\omega_2)=0.10,\; P(\omega_3)=0.20,\; P(\omega_4)=0.20.
$$

Thus

$$
(\Omega, \mathfrak{F}, P) = \left( \{\omega_1,\omega_2,\omega_3,\omega_4\},\; \mathfrak{P}(\Omega),\; P \text{ as above} \right).
$$

---

## 2. $\sigma$-algebra generated by $Y$, $\sigma(Y)$

We have $Y=0$ on $\{\omega_1,\omega_2\}$, $Y=1$ on $\{\omega_3,\omega_4\}$.

Thus

$$
\sigma(Y) = \left\{ \emptyset,\; \{\omega_1,\omega_2\},\; \{\omega_3,\omega_4\},\; \Omega \right\}.
$$

---

## 3. Why $E[X \mid Y]$ is constant on $\{\omega_1,\omega_2\}$ and $\{\omega_3,\omega_4\}$

By definition, $E[X \mid Y]$ is $\sigma(Y)$-measurable.  
A $\sigma(Y)$-measurable random variable must be constant on each atom of $\sigma(Y)$.  
Atoms are $\{\omega_1,\omega_2\}$ and $\{\omega_3,\omega_4\}$.

Thus there exist constants $a,b$ such that

$$
E[X \mid Y](\omega) =
\begin{cases}
a & \omega \in \{\omega_1,\omega_2\},\\
b & \omega \in \{\omega_3,\omega_4\}.
\end{cases}
$$

---

## 4. Compute $E[X \mid Y]$ explicitly

For $A = \{\omega_1,\omega_2\}$:

$$
\int_A E[X \mid Y] \, dP = a \cdot P(\{\omega_1,\omega_2\}) = a \cdot 0.6.
$$

$$
\int_A X \, dP = 0 \cdot 0.50 + 1 \cdot 0.10 = 0.10.
$$

Thus $0.6a = 0.10 \implies a = \frac{1}{6}$.

For $A = \{\omega_3,\omega_4\}$:

$$
\int_A E[X \mid Y] \, dP = b \cdot 0.4 = 0.4b.
$$

$$
\int_A X \, dP = 0 \cdot 0.20 + 1 \cdot 0.20 = 0.20.
$$

Thus $0.4b = 0.20 \implies b = \frac{1}{2}$.

Therefore

$$
E[X \mid Y] = \frac{1}{6} \mathbf{1}_{\{Y=0\}} + \frac{1}{2} \mathbf{1}_{\{Y=1\}}.
$$

---

## 5. Verify the defining property

Already verified in step 4. For completeness:

- $A = \{\omega_1,\omega_2\}$: $\int_A E[X\mid Y] = 0.6 \cdot \frac16 = 0.1 = \int_A X$.
- $A = \{\omega_3,\omega_4\}$: $\int_A E[X\mid Y] = 0.4 \cdot \frac12 = 0.2 = \int_A X$.
- $A = \Omega$: $\int_\Omega E[X\mid Y] = 0.1 + 0.2 = 0.3 = \int_\Omega X$.
- $A = \emptyset$: trivial.

---

## 6. Conditional PMF $p_{X\mid Y}(x \mid y)$

From the joint distribution:

$P(X=1,Y=0) = P(\omega_2)=0.10$, $P(Y=0)=0.6$  
$\implies p_{X\mid Y}(1 \mid 0) = \frac{0.10}{0.6} = \frac16$, $p_{X\mid Y}(0 \mid 0) = \frac{0.50}{0.6} = \frac56$.

$P(X=1,Y=1) = P(\omega_4)=0.20$, $P(Y=1)=0.4$  
$\implies p_{X\mid Y}(1 \mid 1) = \frac{0.20}{0.4} = \frac12$, $p_{X\mid Y}(0 \mid 1) = \frac12$.

Thus

$$
p_{X\mid Y}(x \mid y) =
\begin{cases}
(5/6, 1/6) & \text{if } y=0,\\
(1/2, 1/2) & \text{if } y=1.
\end{cases}
$$

---

## 7. Verify $\mathbb{E}[X \mid Y=y] = \sum_x x \cdot p_{X\mid Y}(x \mid y)$

- $y=0$: $\sum_x x p(x|0) = 0 \cdot \frac56 + 1 \cdot \frac16 = \frac16$.
- $y=1$: $0 \cdot \frac12 + 1 \cdot \frac12 = \frac12$.

Matches $E[X \mid Y=y]$ from step 4.

---

## 8. Mean Square Error (MSE) for $E[X \mid Y]$

By the law of total variance / conditional variance formula:

For binary $X$: $\operatorname{Var}(X \mid Y=y) = E[X\mid Y=y](1 - E[X\mid Y=y])$.

- $Y=0$: $E[X\mid Y=0]=1/6$, $\operatorname{Var}= \frac16 \cdot \frac56 = \frac{5}{36}$, $P(Y=0)=0.6$.
- $Y=1$: $E[X\mid Y=1]=1/2$, $\operatorname{Var}= \frac12 \cdot \frac12 = \frac14$, $P(Y=1)=0.4$.

$$
\text{MSE} = 0.6 \cdot \frac{5}{36} + 0.4 \cdot \frac14
= \frac{3}{36} + 0.1
= \frac{1}{12} + \frac{1}{10}
= \frac{5+6}{60}
= \frac{11}{60}.
$$

$$
\boxed{\frac{11}{60}}
$$

---

## 9. MSE using global average $E[X] = 0.30$

Constant estimate $\hat{X} = 0.30$.

$$
\operatorname{MSE}_{\text{global}} = E[(X - 0.30)^2] = E[X^2] - 2\cdot0.30 E[X] + (0.30)^2.
$$

Since $X$ is $0/1$, $E[X^2] = E[X] = 0.30$.

Thus

$$
\operatorname{MSE}_{\text{global}} = 0.30 - 0.18 + 0.09 = 0.21 = \frac{21}{100}.
$$

**Relationship to Law of Total Variance**:

$$
\operatorname{Var}(X) = \operatorname{Var}(E[X \mid Y]) + E[\operatorname{Var}(X \mid Y)].
$$

Here $\operatorname{Var}(X) = 0.30(0.70) = 0.21$ (matches global MSE).  
$E[\operatorname{Var}(X \mid Y)]$ is MSE using $Y$ = $11/60 \approx 0.18333$.  
Thus $\operatorname{Var}(E[X \mid Y]) = 0.21 - 11/60 = (12.6 - 11)/60?$ Let's compute exactly:

$0.21 = 21/100 = 63/300$, $11/60 = 55/300$, difference $= 8/300 = 2/75$.

So $\operatorname{Var}(E[X \mid Y]) = 2/75$, the reduction in MSE.

---

## 10. Joint distribution and information theory

From $\Omega$:

- $Y=0,X=0$: $\omega_1$, $P=0.5$
- $Y=0,X=1$: $\omega_2$, $P=0.1$
- $Y=1,X=0$: $\omega_3$, $P=0.2$
- $Y=1,X=1$: $\omega_4$, $P=0.2$

Thus $P(X=0)=0.7$, $P(X=1)=0.3$, $P(Y=0)=0.6$, $P(Y=1)=0.4$.

### a) $H(X)$

$$
H(X) = -0.7 \log_2(0.7) - 0.3 \log_2(0.3)
$$

Numerically:  
$0.7 \log_2(0.7) \approx 0.7 \times (-0.51457) = -0.36020$  
$0.3 \log_2(0.3) \approx 0.3 \times (-1.73697) = -0.52109$  
Sum of negatives = $-0.88129$, so $H(X) \approx 0.8813$ bits.

### b) $H(X \mid Y)$

$H(X \mid Y=0) = -\frac56 \log_2(5/6) - \frac16 \log_2(1/6)$  
$\frac56 \log_2(5/6) \approx 0.83333 \times (-0.26303) = -0.21919$  
$\frac16 \log_2(1/6) \approx 0.16667 \times (-2.58496) = -0.43083$  
Sum = $-0.65002$, so $H(X \mid Y=0) \approx 0.6500$ bits.

$H(X \mid Y=1) = -0.5 \log_2(0.5) - 0.5 \log_2(0.5) = 1$ bit.

Thus

$$
H(X \mid Y) = 0.6 \times 0.6500 + 0.4 \times 1 = 0.39 + 0.4 = 0.79 \text{ bits}.
$$

### c) Mutual information $I(X;Y)$

$$
I(X;Y) = H(X) - H(X \mid Y) \approx 0.8813 - 0.79 = 0.0913 \text{ bits}.
$$

### d) Percentage of uncertainty removed

$$
\frac{I(X;Y)}{H(X)} \approx \frac{0.0913}{0.8813} \approx 0.1036 = 10.36\%.
$$

# Q2) Conditional Density of the Multivariate Gaussian

Let $\mathbf{Z}$ be the concatenated random vector

$$
\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}.
$$

$\mathbf{Z}$ has a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

If the covariance matrix is non‑singular (positive definite), the joint density exists:

$$
f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right),
$$

where  

$$
\boldsymbol{\mu} = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix}, \qquad
\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}.
$$

We want to derive the conditional distribution $f_{\mathbf{X} \mid \mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$.

---

## Step 1 – Block matrix inverse (Schur complement)

For a block matrix, the inverse can be written as:

$$
\boldsymbol{\Sigma}^{-1} =
\begin{bmatrix}
\boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\
\boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY}
\end{bmatrix}^{-1}
=
\begin{bmatrix}
\mathbf{A} & \mathbf{B} \\
\mathbf{B}^T & \mathbf{C}
\end{bmatrix},
$$

where  

- $\mathbf{A} = (\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1} \boldsymbol{\Sigma}_{YX})^{-1} =: \boldsymbol{\Sigma}_{X\mid Y}^{-1}$  
- $\mathbf{B} = - \boldsymbol{\Sigma}_{X\mid Y}^{-1} \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1}$  
- $\mathbf{C} = \boldsymbol{\Sigma}_{YY}^{-1} + \boldsymbol{\Sigma}_{YY}^{-1} \boldsymbol{\Sigma}_{YX} \boldsymbol{\Sigma}_{X\mid Y}^{-1} \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1}$

---

## Step 2 – Quadratic form expansion

Let $\mathbf{u} = \mathbf{x} - \boldsymbol{\mu}_X$, $\mathbf{v} = \mathbf{y} - \boldsymbol{\mu}_Y$.

The exponent in the joint density is  

$$
-\frac12 \begin{bmatrix} \mathbf{u}^T & \mathbf{v}^T \end{bmatrix}
\begin{bmatrix} \mathbf{A} & \mathbf{B} \\ \mathbf{B}^T & \mathbf{C} \end{bmatrix}
\begin{bmatrix} \mathbf{u} \\ \mathbf{v} \end{bmatrix}.
$$

Expanding:

$$
= -\frac12 \left( \mathbf{u}^T \mathbf{A} \mathbf{u} + \mathbf{u}^T \mathbf{B} \mathbf{v} + \mathbf{v}^T \mathbf{B}^T \mathbf{u} + \mathbf{v}^T \mathbf{C} \mathbf{v} \right).
$$

Since $\mathbf{u}^T \mathbf{B} \mathbf{v} = \mathbf{v}^T \mathbf{B}^T \mathbf{u}$ (scalar), this becomes  

$$
= -\frac12 \left( \mathbf{u}^T \mathbf{A} \mathbf{u} + 2 \mathbf{u}^T \mathbf{B} \mathbf{v} + \mathbf{v}^T \mathbf{C} \mathbf{v} \right).
$$

---

## Step 3 – Completing the square in $\mathbf{u}$

We want to write this as  

$$
-\frac12 \left( (\mathbf{u} - \mathbf{m})^T \mathbf{A} (\mathbf{u} - \mathbf{m}) \right) + \text{term independent of } \mathbf{u},
$$

where $\mathbf{m}$ will be the conditional mean.

Expand $(\mathbf{u} - \mathbf{m})^T \mathbf{A} (\mathbf{u} - \mathbf{m})$:

$$
= \mathbf{u}^T \mathbf{A} \mathbf{u} - 2 \mathbf{u}^T \mathbf{A} \mathbf{m} + \mathbf{m}^T \mathbf{A} \mathbf{m}.
$$

Comparing with $\mathbf{u}^T \mathbf{A} \mathbf{u} + 2 \mathbf{u}^T \mathbf{B} \mathbf{v}$, we match the linear term:

$$
-2 \mathbf{u}^T \mathbf{A} \mathbf{m} = 2 \mathbf{u}^T \mathbf{B} \mathbf{v} \quad \Rightarrow \quad \mathbf{A} \mathbf{m} = -\mathbf{B} \mathbf{v}.
$$

Thus  

$$
\mathbf{m} = -\mathbf{A}^{-1} \mathbf{B} \mathbf{v}.
$$

But $\mathbf{A}^{-1} = \boldsymbol{\Sigma}_{X\mid Y}$, and $\mathbf{B} = -\boldsymbol{\Sigma}_{X\mid Y}^{-1} \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1}$.

So  

$$
\mathbf{m} = - \boldsymbol{\Sigma}_{X\mid Y} \left( -\boldsymbol{\Sigma}_{X\mid Y}^{-1} \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1} \mathbf{v} \right)
= \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1} \mathbf{v}.
$$

Therefore  

$$
\mathbf{m} = \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y).
$$

---

## Step 4 – Conditional mean and covariance

From the completion of the square:

$$
\mathbf{u}^T \mathbf{A} \mathbf{u} + 2 \mathbf{u}^T \mathbf{B} \mathbf{v} = (\mathbf{u} - \mathbf{m})^T \mathbf{A} (\mathbf{u} - \mathbf{m}) - \mathbf{m}^T \mathbf{A} \mathbf{m}.
$$

The extra term $-\mathbf{m}^T \mathbf{A} \mathbf{m}$ combines with the $\mathbf{v}^T \mathbf{C} \mathbf{v}$ term and with the normalization constant to give the marginal density of $\mathbf{Y}$.

Thus, fixing $\mathbf{y}$, the conditional density of $\mathbf{X} \mid \mathbf{Y} = \mathbf{y}$ is proportional to  

$$
\exp\left( -\frac12 (\mathbf{x} - \boldsymbol{\mu}_X - \mathbf{m})^T \mathbf{A} (\mathbf{x} - \boldsymbol{\mu}_X - \mathbf{m}) \right).
$$

That is:

- Conditional mean:  
  $$
  \boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \mathbf{m} = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y).
  $$

- Conditional covariance (precision matrix is $\mathbf{A}$):  
  $$
  \boldsymbol{\Sigma}_{X\mid Y} = \mathbf{A}^{-1} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1} \boldsymbol{\Sigma}_{YX}.
  $$

---

## Step 5 – Final result

Thus the conditional distribution is multivariate normal:

$$
\boxed{\mathbf{X} \mid \mathbf{Y} = \mathbf{y} \;\sim\; \mathcal{N}\big( \boldsymbol{\mu}^*,\; \boldsymbol{\Sigma}_{X\mid Y} \big)},
$$

with  

$$
\boxed{\boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y)}, \qquad
\boxed{\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1} \boldsymbol{\Sigma}_{YX}}.
$$

The conditional density is therefore:

$$
f_{\mathbf{X} \mid \mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{X\mid Y}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right).
$$

---

**Interpretation**  
- $\boldsymbol{\Sigma}_{X\mid Y}$ is the **Schur complement** of $\boldsymbol{\Sigma}_{YY}$ in $\boldsymbol{\Sigma}$.  
- $\boldsymbol{\mu}^*$ adjusts the mean of $\mathbf{X}$ based on the observed $\mathbf{y}$ via the linear regression coefficient $\boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1}$.

# Q3) Continuous Fault Monitoring of a Spindle Motor

A factory monitors the health of a high-speed CNC spindle. The critical variable is the **Internal Temperature ($X$)**, but because the motor is sealed, they can only observe the **Total Spindle Power ($Y$)**.

**The Setup:**

1. **Hidden Temperature ($X$):**  
   $$X \sim \mathcal{N}(0, 1)$$

2. **Mechanical Load ($Q$):** Independent of $X$:  
   $$Q \sim \mathcal{N}(0, 4)$$

3. **The Measurement ($Y$):**  
   $$Y = X + Q$$

---

## 1. Joint Distribution

**Mean of $Y$:**  
$$E[Y] = E[X] + E[Q] = 0 + 0 = 0.$$

**Variance of $Y$:**  
Since $X$ and $Q$ are independent:  
$$\operatorname{Var}(Y) = \operatorname{Var}(X) + \operatorname{Var}(Q) = 1 + 4 = 5.$$

Thus  
$$\boxed{Y \sim \mathcal{N}(0, 5)}.$$

**Covariance $\operatorname{Cov}(X, Y)$:**  
$$\operatorname{Cov}(X, Y) = \operatorname{Cov}(X, X + Q) = \operatorname{Cov}(X, X) + \operatorname{Cov}(X, Q).$$  
Since $X$ and $Q$ are independent, $\operatorname{Cov}(X, Q)=0$. Thus  
$$\operatorname{Cov}(X, Y) = \operatorname{Var}(X) = 1.$$

So  
$$\boxed{\operatorname{Cov}(X, Y) = 1}.$$

**Joint distribution** $(X, Y)$ is bivariate normal:  
$$
\begin{bmatrix} X \\ Y \end{bmatrix} \sim \mathcal{N}\left( \begin{bmatrix} 0 \\ 0 \end{bmatrix}, \begin{bmatrix} 1 & 1 \\ 1 & 5 \end{bmatrix} \right).
$$

---

## 2. Information Geometry

$X$ is **not** $\sigma(Y)$-measurable because $X$ is not a deterministic function of $Y$.

**Practical explanation:**  
If $Y=5$, it could be because:

- $X=5$ and $Q=0$ (overheating)
- $X=0$ and $Q=5$ (hard material)
- Or any combination $X=2, Q=3$, etc.

Since $Q$ is random, infinitely many $(X, Q)$ pairs give the same $Y$. The engineer cannot uniquely determine $X$ from $Y$ alone.

---

## 3. Conditional Expectation (Best Estimate)

For bivariate normal:

$$
E[X \mid Y = y] = \mu_X + \frac{\operatorname{Cov}(X,Y)}{\operatorname{Var}(Y)} (y - \mu_Y).
$$

Substitute $\mu_X=0$, $\mu_Y=0$, $\operatorname{Cov}(X,Y)=1$, $\operatorname{Var}(Y)=5$:

$$
E[X \mid Y = y] = 0 + \frac{1}{5} (y - 0) = \frac{y}{5}.
$$

Thus  

$$
\boxed{E[X \mid Y] = \frac{1}{5} Y}.
$$

**Signal‑to‑total‑variance ratio:**  

$$
\frac{\operatorname{Cov}(X,Y)}{\operatorname{Var}(Y)} = \frac{1}{5} = \frac{\text{Var}(X)}{\text{Var}(X) + \text{Var}(Q)}.
$$

Interpretation: The estimator “shrinks” $Y$ toward zero by the fraction of total variance due to signal.

---

## 4. Verification of the Residual

Let $\widehat{X} = E[X \mid Y] = Y/5$, and $Z = X - \widehat{X}$.

**Prove $E[Z] = 0$:**  
$$E[Z] = E[X] - E[\widehat{X}] = 0 - \frac{1}{5}E[Y] = 0 - 0 = 0.$$

**Orthogonality principle:** Show $\operatorname{Cov}(Z, Y) = 0$.

First compute $\operatorname{Cov}(\widehat{X}, Y)$:  
$$\operatorname{Cov}(\widehat{X}, Y) = \operatorname{Cov}\left(\frac{Y}{5}, Y\right) = \frac{1}{5} \operatorname{Var}(Y) = \frac{1}{5} \cdot 5 = 1.$$

Now $\operatorname{Cov}(Z, Y) = \operatorname{Cov}(X - \widehat{X}, Y) = \operatorname{Cov}(X, Y) - \operatorname{Cov}(\widehat{X}, Y) = 1 - 1 = 0.$

Thus  

$$
\boxed{E[Z] = 0, \quad \operatorname{Cov}(Z, Y) = 0}.
$$

Residual is uncorrelated with observation — key property of optimal linear (and in this Gaussian case, optimal nonlinear) estimation.

---

## 5. Differential Entropy

**Differential entropy of $X \sim \mathcal{N}(0, \sigma^2)$:**  
$$h(X) = \frac{1}{2} \ln(2\pi e \sigma^2).$$

For $\sigma_X^2 = 1$:  
$$\boxed{h(X) = \frac{1}{2} \ln(2\pi e)}.$$

**Conditional variance $X \mid Y = y$:**  

For bivariate normal:  

$$
\sigma_{X\mid Y}^2 = \sigma_X^2 - \frac{\operatorname{Cov}(X,Y)^2}{\operatorname{Var}(Y)} = 1 - \frac{1^2}{5} = 1 - 0.2 = 0.8.
$$

Note: $\sigma_{X\mid Y}^2$ does **not** depend on $y$.

**Conditional differential entropy:**  
$$h(X \mid Y) = \frac{1}{2} \ln(2\pi e \cdot \sigma_{X\mid Y}^2) = \frac{1}{2} \ln(2\pi e \cdot 0.8).$$

Thus  

$$
\boxed{h(X \mid Y) = \frac{1}{2} \ln(2\pi e \cdot 0.8)}.
$$

---

## 6. Mutual Information

$$I(X; Y) = h(X) - h(X \mid Y) = \frac{1}{2} \ln(2\pi e) - \frac{1}{2} \ln(2\pi e \cdot 0.8).$$

Simplify:  

$$
I(X; Y) = \frac{1}{2} \ln\left( \frac{2\pi e}{2\pi e \cdot 0.8} \right) = \frac{1}{2} \ln\left( \frac{1}{0.8} \right) = \frac{1}{2} \ln(1.25).
$$

Numerically: $\ln(1.25) \approx 0.22314$, so  

$$
\boxed{I(X; Y) \approx 0.11157 \text{ nats}}.
$$

If using bits: $I = \frac{1}{2} \log_2(1.25) \approx 0.1610$ bits.

**Percentage of uncertainty removed:**  
$$ \frac{I(X;Y)}{h(X)} = \frac{\frac{1}{2} \ln(1.25)}{\frac{1}{2} \ln(2\pi e)} = \frac{\ln(1.25)}{\ln(2\pi e)}. $$

$2\pi e \approx 17.079$, $\ln(17.079) \approx 2.837$.  
$\ln(1.25) \approx 0.2231$.  

Fraction $\approx 0.2231 / 2.837 \approx 0.0786 = 7.86\%$.

**If $\operatorname{Var}(Q) = 100$ instead of 4:**  

Then $\operatorname{Var}(Y) = 1 + 100 = 101$, so $\sigma_{X\mid Y}^2 = 1 - \frac{1}{101} = \frac{100}{101} \approx 0.9901$.

Then $I(X;Y) = \frac{1}{2} \ln\left( \frac{1}{100/101} \right) = \frac{1}{2} \ln(1.01) \approx 0.004975$ nats (much smaller).  

Uncertainty removed $\approx 0.004975 / 2.837 \approx 0.175\%$ — almost nothing.

---

## 7. Diagnostic Interpretation

The manager's guess $Y$ is **biased** for $X$ because $E[Y \mid X = x] = x$, but we want the reverse: $E[X \mid Y=y]$.

For Gaussian $X$ and additive independent noise $Q$, the optimal estimator is  

$$
E[X \mid Y] = \frac{\sigma_X^2}{\sigma_X^2 + \sigma_Q^2} Y.
$$

Here $\sigma_X^2=1$, $\sigma_Q^2=4$, so coefficient $= 1/5$.

**Why shrink toward zero?**  
When $\sigma_Q^2$ is large, $Y$ is mostly noise, so the best guess is close to the prior mean (0).  
When $Y$ is extreme, it's more likely due to noise than to a rare extreme $X$.  

Thus $E[X \mid Y]$ “shrinks” $Y$ toward 0 — a **regularization** effect.
